## Loss plot of the FPEnc trained on the reduced dataset

```python
import matplotlib.pyplot as plt
import numpy as np
import json
# Load the train and validation CSV files
# Load the train JSON file
with open('run-train.events-tag-loss_z_mse.json', 'r') as f:
    train_data = json.load(f)

# Extract steps and losses for train data
train_steps = [entry[1] for entry in train_data]
train_losses = [entry[2] for entry in train_data]

# Load the validation JSON file
with open('run-val.events-tag-loss_z_mse.json', 'r') as f:
    val_data = json.load(f)

# Extract steps and losses for validation data
val_steps = [entry[1] for entry in val_data]
val_losses = [entry[2] for entry in val_data]

# Plot the training and validation losses
plt.figure(figsize=(10, 6))
plt.plot(train_steps, train_losses, label='Training', color='blue', linewidth=2)
plt.plot(val_steps, val_losses, label='Validation', color='red', linewidth=2)

# Add horizontal lines for asymptotes
train_asymptote = min(train_losses)
val_asymptote = min(val_losses)
plt.axhline(y=train_asymptote, color='black', linestyle='--', linewidth=2)
plt.axhline(y=val_asymptote, color='black', linestyle='--', linewidth=2)

# Add arrows pointing to the asymptotes and place the values in the middle of the graph
plt.annotate(f'{train_asymptote:.1e}', xy=(train_steps[0], train_asymptote), xytext=(train_steps[len(train_steps)//2], train_asymptote + 0.02),
             arrowprops=dict(facecolor='black', arrowstyle='->'), fontsize=26, family='serif', color='black', verticalalignment='bottom', horizontalalignment='center')

plt.annotate(f'{val_asymptote:.1e}', xy=(val_steps[0], val_asymptote), xytext=(val_steps[len(val_steps)//2], val_asymptote + 0.03),
             arrowprops=dict(facecolor='black', arrowstyle='->'), fontsize=26, family='serif', color='black', verticalalignment='bottom', horizontalalignment='center')

# Add labels, title, legend
plt.xlabel('Iteration ($\\times 10^3$)', fontsize=26, family='serif')
plt.ylabel('Loss', fontsize=26, family='serif')
plt.xticks(fontsize=26, ticks=np.arange(0, 150001, 30000), labels=[f'{int(x/1000)}' for x in np.arange(0, 150001, 30000)], family='serif')
plt.yticks(fontsize=26, family='serif')
plt.legend(prop={'family': 'serif', 'size': 26}, loc='upper right', frameon=False, handlelength=2.5, handleheight=1.5, borderpad=0.5, labelspacing=0.5)

# Save the plot as a high-quality image
plt.tick_params(axis='both', which='both', direction='in', length=6, width=1, colors='black', top=True, right=True)
plt.tight_layout()
plt.savefig('Plot FPEnc Loss Small Dataset.svg', format='svg')

# Show the plot
plt.show()
```
